```python
#converting string into list using module ast
import ast
ast.literal_eval(s1)
#ast.literal_eval checks data type from an unknown source and convert it into its original datatype.

#AST is a module in Python standard library that builds an Abstract Syntax Tree. Automation testing tools and code coverage tools depend on AST to parse the source code and find errors in cod

#The ast.literal_eval method can safely evaluate strings containing Python values from unknown sources without us having to parse the values. However, complex expressions involving indexing or operators cannot be evaluated using this function.
```

```python
#we will merge the "convert" function and ast module
def convert(obj):
  l1=[]
  for i in ast.literal_eval(obj):
    l1.append(i["name"])
  return l1
```

```python
def convert3(obj):
  l1=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      l1.append(i["name"])
      counter+=1
    else:
      break
  return l1
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
  l1=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l1.append(i["name"])
      break
  return l1
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#nltk.stem.porter for steming nd Porter stemmer is the class of nltk library
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
#ps.stem(i) :- if i='love' then output='love', if i='loving' then output='love', if i='loved' then output='love'
#if i='dance' then output='danc', i='dancing' then output='danc',i='danced' then output='danc'
#these example simply signifies that ps.stem(i) creates a core or root word for all similar words, so that models work can be simplified.
```

```python
new_df['tags']=new_df['tags'].apply(stem)
#steming applied to tags
```

```python
#now we will find similarities between tags
#method 1: we can find similarities by matching same words and how many times these words appeared.But this is not a dynamic way of matching tags
#method 2: vectorization :- converting tags into vectors. In this method , inputed movie name's vector is considered as reference and similar movies are recognized by checking which movies' vectors are closest to the reference vector.
#vectorization or text vectorization methods : bag of words,TFIDF,word2vec..etc
```

```python
#we will use " bag of words" method for vectorization. In this method, we concatenate all tags together to orm a large single text paragraph or "large corpus".
#then we sort out 5000 (note: 5000 is not a fixed number, it can be 2000, 3000 etc) most common words from the large text. Then we create a matrix of 5000 x 5000. In 2D vector, these 5000 x 5000 will give combination of two words which will form a vector.
# for example , word number 2 (action) x word number 5000(fiction), this will form a vector between point (0,0) and (2,5000). This vector will be the reference vector and reference words for search. To identify similar movies we just need to check vectors which are closest to the reference vector.
#Above example is for 2 dimensional vector plane. Same logic will work for 5000 dimensional atmosphere.
```

```python
#In vectorization, we will not include "stop words" of the large text. Stop words are the words which are used  to form sentence but does not effect the meaning of the sentence. for example: are,of, and, from, is,a etc.
```

```python
#use sklearn library and CountVectorizer function from sklearn to extract 5000 common words and remove stop words
#CountVectorizer creates a matrix in which each unique word is represented by a column of the matrix, and each text sample from the document is a row in the matrix.
#The value of each cell is nothing but the count of the word in that particular text sample.
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words="english")
```

```python
# we will convert these 5000 words into a numpy array.By default, countvectorizer returns SciPy sparse matrix(Sparse matrices are those matrices that have the most of their elements as zeroes.), thats we will convert numpy array using fit_transform and toarray()
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
#now, we have vectors. next step will be: to calculate the distance between the vectors because more the distanace less the similarity and less distance more similarity.
#we will not calculate euclidean distance (In mathematics, the Euclidean distance between two points in Euclidean space is the length of a line segment between the two points. It can be calculated from the Cartesian coordinates of the points using the Pythagorean theorem, therefore occasionally being called the Pythagorean distance.)
#we will calculate cosine distance i.e theta between 2 vectors.if theta is small, that means both vectors are close.
```

```python
#to measure cosine distance or similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
#we will use above line of code to create general fuction for sorting and giving top five similarity for any movie
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distance=similarity[movie_index]
  movie_list=sorted(list(enumerate(distance)),reverse=True, key=lambda x:x[1])[1:6]

  for i in movie_list:
    print(new_df.iloc[i[0]].title)

```

```python
recommend('Avatar')
```
